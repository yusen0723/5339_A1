```python
df_accredited.columns = df_accredited.columns.str.strip()
df_committed.columns = df_committed.columns.str.strip()
df_probable.columns = df_probable.columns.str.strip()

df_accredited_full = df_accredited.rename(columns={
    "Power station name": "facility_name",
    "State": "state",
    "Installed capacity (MW)": "capacity_mw",
    "Fuel Source (s)": "fuel_source",
    "Accreditation code": "accreditation_code",
    "Postcode": "postcode",
    "Accreditation start date": "accreditation_start_date",
    "Approval date": "approval_date"
}).copy()

df_accredited_full["committed_date"] = pd.NA
df_accredited_full["project_status"] = "accredited"

df_committed_full = df_committed.rename(columns={
    "Project Name": "facility_name",
    "State": "state",
    "MW Capacity": "capacity_mw",
    "Fuel Source": "fuel_source",
    "Committed Date (Month/Year)": "committed_date"
}).copy()

df_committed_full["accreditation_code"] = pd.NA
df_committed_full["postcode"] = pd.NA
df_committed_full["accreditation_start_date"] = pd.NA
df_committed_full["approval_date"] = pd.NA
df_committed_full["project_status"] = "committed"

df_probable_full = df_probable.rename(columns={
    "Probable Projects": "facility_name",
    "State": "state",
    "MW Capacity": "capacity_mw",
    "Fuel Source": "fuel_source"
}).copy()

df_probable_full["accreditation_code"] = pd.NA
df_probable_full["postcode"] = pd.NA
df_probable_full["accreditation_start_date"] = pd.NA
df_probable_full["approval_date"] = pd.NA
df_probable_full["committed_date"] = pd.NA
df_probable_full["project_status"] = "probable"

cer_cols = [
    "facility_name",
    "state",
    "capacity_mw",
    "fuel_source",
    "accreditation_code",
    "postcode",
    "accreditation_start_date",
    "approval_date",
    "committed_date",
    "project_status"
]

df_accredited_full = df_accredited_full.reindex(columns=cer_cols)
df_committed_full = df_committed_full.reindex(columns=cer_cols)
df_probable_full = df_probable_full.reindex(columns=cer_cols)

df_cer = pd.concat(
    [df_accredited_full, df_committed_full, df_probable_full],
    ignore_index=True
)

print(df_cer.shape)
print(df_cer.columns.tolist())
display(df_cer.head())
print(df_cer.info())
```

### Clean NGER

```python
df_nger_clean = df_nger.copy()

df_nger_clean.columns = (
    df_nger_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[^\w]+", "_", regex=True)
    .str.strip("_")
)

text_cols = [
    "reporting_entity",
    "facility_name",
    "type",
    "state",
    "grid_connected",
    "grid",
    "primary_fuel",
    "important_notes",
    "controlling_corporation"
]

for col in text_cols:
    if col in df_nger_clean.columns:
        df_nger_clean[col] = df_nger_clean[col].where(
            df_nger_clean[col].isna(),
            df_nger_clean[col].astype(str).str.strip()
        )

df_nger_clean = df_nger_clean.replace({
    "-": pd.NA,
    "": pd.NA,
    "nan": pd.NA,
    "None": pd.NA
})

numeric_cols = [
    "year",
    "electricity_production_gj",
    "electricity_production_mwh",
    "total_scope_1_emissions_t_co2_e",
    "total_scope_2_emissions_t_co2_e",
    "total_emissions_t_co2_e",
    "emission_intensity_t_co2_e_mwh"
]

for col in numeric_cols:
    if col in df_nger_clean.columns:
        df_nger_clean[col] = (
            df_nger_clean[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .replace("nan", pd.NA)
        )
        df_nger_clean[col] = pd.to_numeric(df_nger_clean[col], errors="coerce")

if "year" in df_nger_clean.columns:
    df_nger_clean["year"] = df_nger_clean["year"].astype("Int64")

if "state" in df_nger_clean.columns:
    df_nger_clean["state"] = df_nger_clean["state"].where(
        df_nger_clean["state"].isna(),
        df_nger_clean["state"].str.upper().str.strip()
    )

df_nger_clean["is_summary"] = df_nger_clean["type"] == "C"

df_nger_clean = df_nger_clean.drop(
    columns=[c for c in ["important_notes"] if c in df_nger_clean.columns]
)

df_nger_clean = df_nger_clean.drop_duplicates(
    subset=["facility_name", "year"]
)

print(df_nger_clean.shape)
print(df_nger_clean.columns.tolist())
print(df_nger_clean.isna().sum())
display(df_nger_clean.head())
print(df_nger_clean.info())

df_nger_clean.to_csv("cleaned/cleaned_nger.csv", index=False)
```

### Clean CER

```python
df_cer_clean = df_cer.copy()

# 1. 列名统一（保险）
df_cer_clean.columns = (
    df_cer_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[^\w]+", "_", regex=True)
    .str.strip("_")
)

# 2. 文本清洗
text_cols = ["facility_name", "state", "fuel_source", "project_status"]

for col in text_cols:
    if col in df_cer_clean.columns:
        df_cer_clean[col] = df_cer_clean[col].where(
            df_cer_clean[col].isna(),
            df_cer_clean[col].astype(str).str.strip()
        )

# 3. state 统一
df_cer_clean["state"] = df_cer_clean["state"].str.upper().str.strip()

# 4. capacity 转 numeric
df_cer_clean["capacity_mw"] = (
    df_cer_clean["capacity_mw"]
    .astype(str)
    .str.replace(",", "", regex=False)
)
df_cer_clean["capacity_mw"] = pd.to_numeric(df_cer_clean["capacity_mw"], errors="coerce")

# 5. 日期统一（自动识别两种格式）
date_cols = [
    "accreditation_start_date",
    "approval_date",
    "committed_date"
]

for col in date_cols:
    if col in df_cer_clean.columns:
        df_cer_clean[col] = pd.to_datetime(df_cer_clean[col], errors="coerce")

# 6. 删除明显没用列（可选）
df_cer_clean = df_cer_clean.drop(columns=[
    c for c in ["postcode"] if c in df_cer_clean.columns
])

# 7. 去重
df_cer_clean = df_cer_clean.drop_duplicates()

# 查看结果
print(df_cer_clean.shape)
print(df_cer_clean.dtypes)
display(df_cer_clean.head())

df_cer_clean.to_csv("cleaned/cleaned_cer.csv", index=False)
```

### Clean AbS

```python
df_abs_clean = df_abs.copy()

df_abs_clean.columns = (
    df_abs_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(r"[^\w]+", "_", regex=True)
    .str.strip("_")
)

drop_cols = [
    "dataflow",
    "regiontype_region_type",
    "frequency_frequency",
    "unit_mult_unit_of_multiplier",
    "obs_status_observation_status",
    "obs_comment_observation_comment"
]

df_abs_clean = df_abs_clean.drop(columns=[c for c in drop_cols if c in df_abs_clean.columns])

df_abs_clean["state"] = (
    df_abs_clean["asgs_2021_region"]
    .astype(str)
    .str.replace(r"^\d+:\s*", "", regex=True)
    .str.strip()
)

state_map = {
    "New South Wales": "NSW",
    "Victoria": "VIC",
    "Queensland": "QLD",
    "South Australia": "SA",
    "Western Australia": "WA",
    "Tasmania": "TAS",
    "Northern Territory": "NT",
    "Australian Capital Territory": "ACT",
    "Other Territories": "OT"
}

df_abs_clean["state"] = df_abs_clean["state"].map(state_map)

df_abs_clean["year"] = pd.to_numeric(
    df_abs_clean["time_period_time_period"],
    errors="coerce"
).astype("Int64")

df_abs_clean["obs_value"] = (
    df_abs_clean["obs_value"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .replace("nan", pd.NA)
)
df_abs_clean["obs_value"] = pd.to_numeric(df_abs_clean["obs_value"], errors="coerce")

df_abs_clean = df_abs_clean.drop(columns=[
    c for c in ["asgs_2021_region"] if c in df_abs_clean.columns
])

print(df_abs_clean.shape)
print(df_abs_clean.columns.tolist())
print(df_abs_clean["state"].value_counts(dropna=False))
display(df_abs_clean.head())

df_abs_clean.to_csv("cleaned/cleaned_abs.csv", index=False)
```

### join data

```python
# ============================================================
# Merge latest cleaned tables into one big denormalised table
# Use:
#   - df_nger_clean
#   - df_cer_clean
#   - df_abs_wide
# Keep all attributes, do not worry about table size
# ============================================================

import pandas as pd

# -----------------------------
# Step 1: copy cleaned tables
# -----------------------------
nger_big = df_nger_clean.copy()
cer_big = df_cer_clean.copy()
abs_big = df_abs_wide.copy()

# -----------------------------
# Step 2: helper to find columns safely
# -----------------------------
def find_first_existing(df, candidates):
    for c in candidates:
        if c in df.columns:
            return c
    raise KeyError(f"None of these columns exist: {candidates}")

# -----------------------------
# Step 3: identify key columns
# -----------------------------
# NGER
nger_state_col = find_first_existing(nger_big, ["state", "State"])
nger_year_col = find_first_existing(nger_big, ["year", "Year"])

# CER
cer_state_col = find_first_existing(cer_big, ["state", "State"])

# ABS wide
abs_state_col = find_first_existing(abs_big, ["state", "State"])
abs_year_col = find_first_existing(abs_big, ["year", "Year"])

# -----------------------------
# Step 4: build merge keys
# -----------------------------
# NGER keys
nger_big["merge_state"] = nger_big[nger_state_col].astype(str).str.upper().str.strip()
nger_big["merge_year"] = pd.to_numeric(nger_big[nger_year_col], errors="coerce").astype("Int64")

# CER keys
cer_big["merge_state"] = cer_big[cer_state_col].astype(str).str.upper().str.strip()

# build CER merge_year from available date columns
cer_big["merge_year"] = pd.NA

if "committed_date" in cer_big.columns:
    cer_big["merge_year"] = pd.to_datetime(cer_big["committed_date"], errors="coerce").dt.year

if "approval_date" in cer_big.columns:
    cer_big["merge_year"] = cer_big["merge_year"].fillna(
        pd.to_datetime(cer_big["approval_date"], errors="coerce").dt.year
    )

if "accreditation_start_date" in cer_big.columns:
    cer_big["merge_year"] = cer_big["merge_year"].fillna(
        pd.to_datetime(cer_big["accreditation_start_date"], errors="coerce").dt.year
    )

cer_big["merge_year"] = pd.to_numeric(cer_big["merge_year"], errors="coerce").astype("Int64")

# ABS wide keys
abs_big["merge_state"] = abs_big[abs_state_col].astype(str).str.upper().str.strip()
abs_big["merge_year"] = pd.to_numeric(abs_big[abs_year_col], errors="coerce").astype("Int64")

# -----------------------------
# Step 5: add row ids
# -----------------------------
nger_big["nger_row_id"] = range(1, len(nger_big) + 1)
cer_big["cer_row_id"] = range(1, len(cer_big) + 1)
abs_big["abs_row_id"] = range(1, len(abs_big) + 1)

# -----------------------------
# Step 6: prefix columns to preserve all attributes clearly
# -----------------------------
def prefix_non_key_columns(df, prefix, key_cols):
    df = df.copy()
    rename_map = {}
    for col in df.columns:
        if col not in key_cols:
            rename_map[col] = f"{prefix}_{col}"
    return df.rename(columns=rename_map)

key_cols = ["merge_state", "merge_year"]

nger_pref = prefix_non_key_columns(nger_big, "nger", key_cols)
cer_pref  = prefix_non_key_columns(cer_big,  "cer",  key_cols)
abs_pref  = prefix_non_key_columns(abs_big,  "abs",  key_cols)

# -----------------------------
# Step 7: full outer join NGER + CER
# -----------------------------
df_big_1 = nger_pref.merge(
    cer_pref,
    on=["merge_state", "merge_year"],
    how="outer"
)

print("After NGER + CER merge:", df_big_1.shape)

# -----------------------------
# Step 8: full outer join with ABS wide
# -----------------------------
df_one_big_table = df_big_1.merge(
    abs_pref,
    on=["merge_state", "merge_year"],
    how="left"
)

print("Final one big table shape:", df_one_big_table.shape)

# -----------------------------
# Step 9: rename merge keys back to normal names
# -----------------------------
df_one_big_table = df_one_big_table.rename(columns={
    "merge_state": "state",
    "merge_year": "year"
})

# Put keys at the front
front_cols = ["state", "year"]
other_cols = [c for c in df_one_big_table.columns if c not in front_cols]
df_one_big_table = df_one_big_table[front_cols + other_cols]

display(df_one_big_table.head())
```

```python
api_key = "AIzaSyDD1Vvj_fzwINPcGqiNASVp98toVDDhC_c"

def query_facility(facility, state, max_retries=3):
    """
    Query Google Geocoding API v4 for one facility.
    Returns a dictionary instead of only (lat, lng).
    """
    address = f"{facility}, {state}, Australia"
    encoded_address = quote(address, safe="")
    url = f"https://geocode.googleapis.com/v4/geocode/address/{encoded_address}"

    headers = {
        "X-Goog-Api-Key": api_key
    }

    for attempt in range(max_retries):
        try:
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()
            data = response.json()

            if "results" in data and len(data["results"]) > 0:
                top = data["results"][0]
                location = top["location"]

                return {
                    "geo_query": address,
                    "latitude": location.get("latitude"),
                    "longitude": location.get("longitude"),
                    "place_id": top.get("placeId"),
                    "formatted_address": top.get("formattedAddress"),
                    "geocode_status": "OK"
                }

            return {
                "geo_query": address,
                "latitude": None,
                "longitude": None,
                "place_id": None,
                "formatted_address": None,
                "geocode_status": "NO_RESULT"
            }

        except requests.exceptions.RequestException as e:
            print(f"⚠️ Attempt {attempt + 1} failed: {e}")
            time.sleep(0.5)

    return {
        "geo_query": address,
        "latitude": None,
        "longitude": None,
        "place_id": None,
        "formatted_address": None,
        "geocode_status": "REQUEST_ERROR"
    }
```

```python
# ============================================================
# Store cleaned / augmented data into DuckDB with spatial extension
# Create source tables + one big table named: full_table
# ============================================================

import duckdb
import pandas as pd
import os

# -----------------------------
# 0. make output folder
# -----------------------------
os.makedirs("database", exist_ok=True)

# -----------------------------
# 1. connect DuckDB
# -----------------------------
db_path = "database/energy_project.duckdb"
con = duckdb.connect(db_path)

# -----------------------------
# 2. install + load spatial extension
# -----------------------------
con.execute("INSTALL spatial;")
con.execute("LOAD spatial;")

# -----------------------------
# 3. select dataframes
# -----------------------------
if "df_nger_augmented" in globals():
    nger_df = df_nger_augmented.copy()
elif "df_nger_clean" in globals():
    nger_df = df_nger_clean.copy()
elif "df_nger" in globals():
    nger_df = df_nger.copy()
else:
    raise ValueError("No NGER dataframe found.")

if "df_cer_clean" in globals():
    cer_df = df_cer_clean.copy()
elif "df_cer" in globals():
    cer_df = df_cer.copy()
else:
    raise ValueError("No CER dataframe found.")

if "df_abs_wide" in globals():
    abs_df = df_abs_wide.copy()
elif "df_abs_clean" in globals():
    abs_df = df_abs_clean.copy()
elif "df_abs" in globals():
    abs_df = df_abs.copy()
else:
    raise ValueError("No ABS dataframe found.")

# -----------------------------
# 4. clean column names
# -----------------------------
def clean_columns(df):
    df = df.copy()
    df.columns = (
        pd.Index(df.columns)
        .str.strip()
        .str.lower()
        .str.replace(r"[^\w]+", "_", regex=True)
        .str.strip("_")
    )
    return df

nger_df = clean_columns(nger_df)
cer_df = clean_columns(cer_df)
abs_df = clean_columns(abs_df)

# -----------------------------
# 5. standardise state/year
# -----------------------------
def std(df):
    df = df.copy()
    if "state" in df.columns:
        df["state"] = df["state"].astype(str).str.upper().str.strip()
    if "year" in df.columns:
        df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")
    return df

nger_df = std(nger_df)
cer_df = std(cer_df)
abs_df = std(abs_df)

# -----------------------------
# 6. register
# -----------------------------
con.register("nger_df_view", nger_df)
con.register("cer_df_view", cer_df)
con.register("abs_df_view", abs_df)

# -----------------------------
# 7. drop tables
# -----------------------------
for tbl in ["nger_facilities", "cer_projects", "abs_state_year",
            "full_table", "nger_tmp", "cer_tmp", "abs_tmp"]:
    con.execute(f"DROP TABLE IF EXISTS {tbl};")

# -----------------------------
# 8. create source tables
# -----------------------------
con.execute("CREATE TABLE nger_facilities AS SELECT * FROM nger_df_view;")
con.execute("CREATE TABLE cer_projects AS SELECT * FROM cer_df_view;")
con.execute("CREATE TABLE abs_state_year AS SELECT * FROM abs_df_view;")

# -----------------------------
# 9. add id
# -----------------------------
con.execute("""
CREATE OR REPLACE TABLE nger_facilities AS
SELECT row_number() OVER () AS nger_id, * FROM nger_facilities;
""")

con.execute("""
CREATE OR REPLACE TABLE cer_projects AS
SELECT row_number() OVER () AS cer_id, * FROM cer_projects;
""")

con.execute("""
CREATE OR REPLACE TABLE abs_state_year AS
SELECT row_number() OVER () AS abs_id, * FROM abs_state_year;
""")

# -----------------------------
# 10. spatial column for NGER
# -----------------------------
cols = [r[1] for r in con.execute("PRAGMA table_info('nger_facilities')").fetchall()]

if "latitude" in cols and "longitude" in cols:
    con.execute("""
    CREATE OR REPLACE TABLE nger_facilities AS
    SELECT *,
    CASE WHEN longitude IS NOT NULL AND latitude IS NOT NULL
         THEN ST_Point(CAST(longitude AS DOUBLE), CAST(latitude AS DOUBLE))
         ELSE NULL END AS geom
    FROM nger_facilities;
    """)

# -----------------------------
# 11. temp tables (keys)
# -----------------------------
con.execute("""
CREATE TABLE nger_tmp AS
SELECT *, UPPER(TRIM(state)) AS merge_state,
       CAST(year AS INTEGER) AS merge_year
FROM nger_facilities;
""")

con.execute("""
CREATE TABLE cer_tmp AS
SELECT *,
    UPPER(TRIM(state)) AS merge_state,
    COALESCE(
        YEAR(TRY_CAST(committed_date AS DATE)),
        YEAR(TRY_CAST(approval_date AS DATE)),
        YEAR(TRY_CAST(accreditation_start_date AS DATE))
    ) AS merge_year
FROM cer_projects;
""")

con.execute("""
CREATE TABLE abs_tmp AS
SELECT *, UPPER(TRIM(state)) AS merge_state,
       CAST(year AS INTEGER) AS merge_year
FROM abs_state_year;
""")

# -----------------------------
# 12. ONE BIG TABLE
# -----------------------------
con.execute("""
CREATE TABLE full_table AS
SELECT
    COALESCE(n.merge_state, c.merge_state, a.merge_state) AS state,
    COALESCE(n.merge_year, c.merge_year, a.merge_year) AS year,

    n.* EXCLUDE (merge_state, merge_year),
    c.* EXCLUDE (merge_state, merge_year),
    a.* EXCLUDE (merge_state, merge_year)

FROM nger_tmp n
FULL OUTER JOIN cer_tmp c
    ON n.merge_state = c.merge_state
   AND n.merge_year = c.merge_year

LEFT JOIN abs_tmp a
    ON COALESCE(n.merge_state, c.merge_state) = a.merge_state
   AND COALESCE(n.merge_year, c.merge_year) = a.merge_year;
""")

# -----------------------------
# 13. add id
# -----------------------------
con.execute("""
CREATE OR REPLACE TABLE full_table AS
SELECT row_number() OVER () AS full_id, * FROM full_table;
""")

# -----------------------------
# 14. spatial for full_table
# -----------------------------
cols = [r[1] for r in con.execute("PRAGMA table_info('full_table')").fetchall()]

if "longitude" in cols and "latitude" in cols:
    con.execute("""
    CREATE OR REPLACE TABLE full_table AS
    SELECT *,
    CASE WHEN longitude IS NOT NULL AND latitude IS NOT NULL
         THEN ST_Point(CAST(longitude AS DOUBLE), CAST(latitude AS DOUBLE))
         ELSE NULL END AS geom
    FROM full_table;
    """)

# -----------------------------
# 15. checks
# -----------------------------
print("=== tables ===")
print(con.execute("SHOW TABLES").fetchdf())

print("\n=== row counts ===")
print("nger:", con.execute("SELECT COUNT(*) FROM nger_facilities").fetchone()[0])
print("cer :", con.execute("SELECT COUNT(*) FROM cer_projects").fetchone()[0])
print("abs :", con.execute("SELECT COUNT(*) FROM abs_state_year").fetchone()[0])
print("full:", con.execute("SELECT COUNT(*) FROM full_table").fetchone()[0])

print("\n=== preview ===")
display(con.execute("SELECT * FROM full_table LIMIT 5").fetchdf())

```
